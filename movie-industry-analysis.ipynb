```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjusts the configuration of the plots we will create

# Now we can read in the data

movies_df = pd.read_csv('../input/movies/movies.csv')
```

```python
# Now let's take a look at the data

movies_df
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in movies_df.columns:
    pct_missing = np.mean(movies_df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing)))
```

```python
for col in movies_df.columns:
    pct_missing = np.mean(movies_df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
plt.scatter(x=movies_df['budget'], y=movies_df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# plot budget vs gross using seaborn

sns.regplot(x='budget',y='gross', data=movies_df)
# scatter_kws={color:"purple"}, line_kws={color:"blue"}
```

```python
# Correlation Matrix between all numeric columns

movies_df.corr(method ='pearson')
```

```python
movies_df.corr(method ='kendall')
```

```python
movies_df.corr(method ='spearman')
```

```python
correlation_matrix = movies_df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Let's add non-numeric values to the correlation matrix such as company
# Using factorize - this assigns a random numeric value for each unique categorical value

movies_df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = movies_df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = movies_df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```

**QUICK ANALYSIS**

```python
# Let's look at the top 15 compaies by gross revenue

CompanyGrossSum = movies_df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```

```python
movies_df['Year'] = movies_df['released'].astype(str).str[:4]
movies_df
```

```python
movies_df.groupby(['company', 'year'])[["gross"]].sum()
```

```python
CompanyGrossSum = movies_df.groupby(['company', 'year'])[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values(['gross','company','year'], ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```
